```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
train = pd.read_csv('../data/unprocessed_data.csv')
train
```

```python
train.rename(columns={"sr": "Snoring Range", "rr" : "Respiration Rate", "t" : "Body Temperature", 'lm' : "Limb Movment Rate", "bo" : "Blood Oxygen", "rem" : "Eye Movement", "hs" : "Number Of Hours Sleep", "hr": "Heart Rate", "sl": "Stress Level"}, inplace=True)
train
```

```python
train.isnull().sum()
```

```python
train['Snoring Range'].fillna(train['Snoring Range'].mean(), inplace=True)
train['Respiration Rate'].fillna(train['Respiration Rate'].mean(), inplace=True)
train['Body Temperature'].fillna(train['Body Temperature'].mean(), inplace=True)
train['Limb Movment Rate'].fillna(train['Limb Movment Rate'].mean(), inplace=True)
train['Blood Oxygen'].fillna(train['Blood Oxygen'].mean(), inplace=True)
train['Eye Movement'].fillna(train['Eye Movement'].mean(), inplace=True)
train['Number Of Hours Sleep'].fillna(train['Number Of Hours Sleep'].mean(), inplace=True)
train['Heart Rate'].fillna(train['Heart Rate'].mean(), inplace=True)
train['Stress Level'].fillna(train['Stress Level'].median(), inplace=True)
train.isnull().sum()
```

```python
train.to_csv("../data/Stress_dataset.csv", index=False)
```
